```python
import numpy as np

# Sigmoid and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
    return x * (1 - x)

# MLP with 1 hidden layer (2 neurons)
def mlp_train(X, y, lr=0.5, epochs=10000):
    np.random.seed(0)
    # Random weights: input→hidden, hidden→output
    wh = np.random.uniform(size=(2, 2))  # 2 inputs, 2 hidden
    wo = np.random.uniform(size=(2, 1))  # 2 hidden, 1 output

    for _ in range(epochs):
        # Forward pass
        hidden = sigmoid(np.dot(X, wh))
        output = sigmoid(np.dot(hidden, wo))

        # Backpropagation
        err_out    = y - output
        delta_out  = err_out * sigmoid_deriv(output)

        err_hid    = delta_out.dot(wo.T)
        delta_hid  = err_hid * sigmoid_deriv(hidden)

        # Update weights
        wo += lr * hidden.T.dot(delta_out)
        wh += lr * X.T.dot(delta_hid)

    return wh, wo

def mlp_predict(X, wh, wo):
    hidden = sigmoid(np.dot(X, wh))
    output = sigmoid(np.dot(hidden, wo))
    return (output >= 0.5).astype(int)

print("Functions defined.")
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])

# AND-NOT: A AND (NOT B)
y_andnot = np.array([[0],[0],[1],[0]])
wh, wo = mlp_train(X, y_andnot)
pred = mlp_predict(X, wh, wo)

print("--- AND-NOT ---")
print(" A  B | Target | Output")
for i in range(4):
    print(f" {X[i,0]}  {X[i,1]} |   {y_andnot[i,0]}    |   {pred[i,0]}")

# XOR
y_xor = np.array([[0],[1],[1],[0]])
wh, wo = mlp_train(X, y_xor)
pred = mlp_predict(X, wh, wo)

print("\n--- XOR ---")
print(" A  B | Target | Output")
for i in range(4):
    print(f" {X[i,0]}  {X[i,1]} |   {y_xor[i,0]}    |   {pred[i,0]}")
```
