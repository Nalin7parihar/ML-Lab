```python
from sklearn.neural_network import MLPClassifier

def mlp_logic_gate(gate):
    X = [[0,0],[0,1],[1,0],[1,1]]

    if gate == 'AND-NOT':
        y = [0,0,1,0]      # A AND (NOT B)
    elif gate == 'XOR':
        y = [0,1,1,0]
    else:
        print("Invalid gate")
        return

    model = MLPClassifier(
        hidden_layer_sizes=(2,),
        max_iter=10000,
        random_state=0
    )

    model.fit(X, y)

    print(f"\n{gate} Gate")
    print("Input -> Output")

    for x in X:
        print(f"{x} -> {model.predict([x])[0]}")

mlp_logic_gate('AND-NOT')
mlp_logic_gate('XOR')
```
