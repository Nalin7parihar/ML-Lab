```python
# Q1: Visualize the n-dimensional data using Scatter plots.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("ToyotaCorolla.csv")

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['KM'], data['Price'], c=data['HP'], cmap='jet', alpha=0.6, edgecolors='k', s=30)
plt.colorbar(label='HP')
plt.xlabel('KM')
plt.ylabel('Price')
plt.title('Scatter Plot: KM vs Price (colored by HP)')
plt.show()
```

```python
# Write a program to implement Hill Climbing Algorithm.

import numpy as np

def objective(x):
    """Objective function to maximize: -(x^2)"""
    return -(x ** 2)

def hill_climbing(start, step_size=0.1, max_iter=1000):
    current = start
    current_val = objective(current)

    for i in range(max_iter):
        left = current - step_size
        right = current + step_size

        left_val = objective(left)
        right_val = objective(right)

        if left_val > current_val and left_val >= right_val:
            current, current_val = left, left_val
        elif right_val > current_val:
            current, current_val = right, right_val
        else:
            break

    return current, current_val

# Run hill climbing from a random start
start = np.random.uniform(-10, 10)
best_x, best_val = hill_climbing(start)
print(f"Starting point: {start:.4f}")
print(f"Best solution:  x = {best_x:.4f}, f(x) = {best_val:.4f}")
```
